```python
txt = 'but soft what light in yonder window breaks'
words = txt.split()
print(words)

t = list()
for word in words:
   t.append((len(word), word)) # pack the tuple (decorate)

t.sort(reverse=True) # sort the tuple (sort)

res = list()    # print out the result (undecorate)
for length, word in t:
    res.append(word)

print(res)

```

```python
"""Searches for a list in a file """

"""Imports a text file """
"""def open_txt():
    filename = "mbox.txt"
    file = open(filename, 'r')
    print("Opened " + filename + ".")
    return file
    return filename"""
        
def addresses():
    filename = "mbox.txt"
    file = open(filename, 'r')
    print("Opened " + filename + ".")
  
    add_lst = []
    for line in file:
        if "From" in line:
            a = line.split()
            if len(a) > 2:
                add_lst.append(a[1])
        return add_lst
    
def top_commits():
    addresses()
    num_lst = []
    for item in range(len(add_lst)):
        email = add_lst[item]
        commit = add_lst.count(add_lst[item])
        num = commit, email
        num_lst.append(num)
    num_lst.sort(reverse=True)

    return(num_lst[0])
```

```python
top_commits()
```
